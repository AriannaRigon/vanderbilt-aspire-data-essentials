```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
```

```python
 hi_df['healthy_people_objective'] =  pop_and_cancer['healthy_people_objective'] 
```

```python
hi_df.head(5)
```

```python
hi_df_dum = pd.get_dummies (hi_df)
```

```python
hi_df_dum.head (5)
```

```python
x = hi_df_dum[['median_age_2017', 'normalized_hospice_rate', 'healthy_people_objective_No', 'healthy_people_objective_Yes']]
```

```python
x.head (5)
```

```python
y = hi_df_dum [['5year_death_trend']]
```

```python
y.head (5)
```

```python
x_train, x_test, y_train, y_test = train_test_split (x, y, test_size=.2, random_state=45)
```

```python
regr = linear_model.LinearRegression()
regr.fit (x_train, y_train)
```

```python
y_pred = regr.predict(x_test)
```

```python
print ('Coefficients: \n', regr.coef_)
```

```python
print ("Mean squared error: %.2f" % mean_squared_error (y_test, y_pred))
```

```python
print ("Variance score: %.2f" % r2_score (y_test, y_pred))
```

```python
plt.scatter(regr.predict(x_train), regr.predict(x_train) - y_train, c='b', s=40, alpha=0.5)
plt.scatter(regr.predict(x_test), regr.predict(x_test) - y_test, c='g', s=40)
plt.hlines(y=0, xmin=-1.5, xmax=1.5)
plt.title('Residuals using train (blue) and test (green)')
plt.ylabel('residuals')
```
